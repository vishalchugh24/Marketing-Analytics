```python
# libraries
import os
import pandas as pd
from cryptography.fernet import Fernet
```

```python
# explore the data
trends_quanlitative = pd.read_csv('trendmart_customer_persona_qualitative.csv')
trends_quantitative = pd.read_csv('trendmart_customer_persona_quantitative.csv')
```

```python
# head of quanlitative
trends_quanlitative.head()
```

```python
# head of quantitative
trends_quantitative.head()
```

```python
# merge data
trends_mart = pd.merge(trends_quanlitative, trends_quantitative, on = 'Customer_ID', how = 'inner')
trends_mart.head()
```

```python
# trends_mart to csv
trends_mart.to_csv('trends_mart.csv', index = False)
```

```python
# generate key
key = Fernet.generate_key()
```

```python
# store it in file
with open('filekey.key', 'wb') as f:
  f.write(key)
```

```python
# read the key
with open('filekey.key', 'rb') as f:
  key = f.read()
```

```python
# store it in fernet
fernet = Fernet(key)
```

```python
# encryption
encryption = fernet.encrypt(data)
```
