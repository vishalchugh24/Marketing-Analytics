```python
# import libraries
import os
import pandas as pd
from cryptography.fernet import Fernet
```

```python
# opwn the file key
with open('filekey.key', 'rb') as f:
  key = f.read()
```

```python
# store key in the fernet variable
fernet = Fernet(key)
```

```python
# open the file
with open('trends_mart.csv', 'rb') as f:
  data = f.read()
```

```python
# Decrypt
decryption = fernet.decrypt(data)
```

```python
# overwrite the data with decryption
with open('trends_mart.csv', 'wb') as f:
  f.write(decryption)
```

```python
# Group customers by Location (Urban/Rural) and find the average Annual Income.
trends_mart.groupby('Location')['Annual Income'].mean().round(2)
```

```python
# Find the most common Communication Preference for Urban customers.
trends_mart[trends_mart['Location'] == 'Urban']['Communication Preference'].mode()
```

```python
# Group customers by Technology Usage and calculate the average Online Purchase Frequency.
trends_mart.groupby('Technology Usage')['Online Purchase Frequency'].mean().round(2)
```

```python
# Find the number of customers whose Goals and Aspirations is "Better health and fitness".
trends_mart[(trends_mart['Goals and Aspirations'] == 'Better health and fitness')]['Customer_ID'].count()
```

```python
# Group customers by Lifestyle and values and find the average Annual Income.
trends_mart.groupby(['Lifestyle and Values'])['Annual Income'].mean().round(2)
```

```python
# Find the most common Decision-Making trigger for customers with High Technology Usage.
trends_mart[trends_mart['Technology Usage'] == 'High']['Decision Making triggers'].mode()
```
