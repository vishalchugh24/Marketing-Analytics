```python
# import the library
from cryptography.fernet import Fernet
```

```python
# opening the filekey
with open('filekey.key' , 'rb') as f:
  key = f.read()
```

```python
# store the file key in an object
fernet = Fernet(key)
```

```python
# open the files
with open('trendmart_customer_persona_quantitative.csv' , 'rb') as f:
  encrypted_quant_data = f.read()

with open('trendmart_customer_persona_qualitative.csv', 'rb') as f:
  encrypted_quali_data = f.read()
```

```python
# Decrypt Data
decrypt_quant_data = fernet.decrypt(encrypted_quant_data)
decrypt_quali_data = fernet.decrypt(encrypted_quali_data)
```

```python
# overwrite my data
with open('trendmart_customer_persona_quantitative.csv', 'wb') as f:
  f.write(decrypt_quant_data)
with open('trendmart_customer_persona_qualitative.csv', 'wb') as f:
  f.write(decrypt_quali_data)
```

Group customers by Location (Urban/Rural) and find the average Annual Income.

```python
# lets see the column names
trend_mart_data.columns
```

```python
# group customer by location and find out the mean of annual income
mean_customer_loc = trend_mart_data.groupby('Location')['Annual Income'].mean()
round(mean_customer_loc,2)
```

Find the most common Communication Preference for Urban customers.

```python
# Location is Urban and most communication preference is mode
communication_pref_urban = trend_mart_data[trend_mart_data['Location'] == 'Urban']['Communication Preference'].mode()
communication_pref_urban
```

```python
# see all the unique values in Communicaton Preference
trend_mart_data['Communication Preference'].unique()
```

```python
# Group customers by Technology Usage and calculate the average Online Purchase Frequency.
mean_tech = trend_mart_data.groupby('Technology Usage')['Online Purchase Frequency'].mean()
round(mean_tech,2)
```

```python
# Find the number of customers whose Goals and Aspirations is "Better health and fitness".
trend_mart_data[trend_mart_data['Goals and Aspirations'] == 'Better health and fitness'].shape[0]
```

```python
# Group customers by Lifestyle and values and find the average Annual Income.
mean_lifestyle = trend_mart_data.groupby('Lifestyle and Values')['Annual Income'].mean()
round(mean_lifestyle,2)
```

```python
# Find the most common Decision-Making trigger for customers with High Technology Usage.
decision_making_tech = trend_mart_data[trend_mart_data['Technology Usage'] == 'High']['Decision Making triggers'].mode()
decision_making_tech
```

```python
# customers who are in Urban areas whose income is greater than 1,000,000 and technology usage is very high
trend_mart_data[(trend_mart_data['Location'] == "Urban") & (trend_mart_data["Annual Income"] > 1000000) & (trend_mart_data["Technology Usage"] == 'High')][["Name", "Age", "Occupation", "Goals and Aspirations"]]
```

```python
# Count how many customers have "Discounts and offers" as their Decision-Making trigger and shop online more than 5 times
trend_mart_data[(trend_mart_data['Decision Making triggers'] == 'Discounts and offers')  & (trend_mart_data['Online Purchase Frequency'] > 5)][["Name", "Age", "Occupation", "Goals and Aspirations"]]
```

```python
# Calculate Customer Lifetime Value
trend_mart_data['clv_online'] = trend_mart_data['Average Order Value'] * trend_mart_data['Online Purchase Frequency'] * trend_mart_data['Customer Lifespan Years']
```

```python

sorted_columns = trend_mart_data[['Customer_ID', 'clv_online']].sort_values(by='clv_online', ascending= False)
sorted_columns.head()
```

```python
# ROMI
trend_mart_data['ROMI'] = ((trend_mart_data['Revenue Generated'] - trend_mart_data['Marketing Spend']) / trend_mart_data['Marketing Spend']) * 100
```

```python
sorted_columns = trend_mart_data[['Customer_ID', 'ROMI']].sort_values(by='ROMI', ascending= True)
sorted_columns.head()
```

```python
# Net Promotor Score -> Promoters, Detractors and Passives

# Count Promoters (NPS_Score >= 9)
promoters_count = trend_mart_data[trend_mart_data['NPS Score'] >= 9].shape[0]
print(f"The Count of Promoters are: {promoters_count}")

# Count Passives (NPS_Score >= 7 and NPS_Score <= 8)
passives_count = trend_mart_data[(trend_mart_data['NPS Score'] >= 7) & (trend_mart_data['NPS Score'] <= 8)].shape[0]
print(f"The count of Passives are: {passives_count}")

# Count Detractors
detractors_count= trend_mart_data[(trend_mart_data['NPS Score'] <=6) & (trend_mart_data['NPS Score'] >= 0)].shape[0]
print(f"The count of Detractors are: {detractors_count}")
```
