```python
# import libraries
from cryptography.fernet import Fernet
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
import warnings
import copy
```

```python
# open the key file
with open('filekey.key', 'rb') as f:
    key = f.read()
```

```python
# Create the object
fernet = Fernet(key)
```

```python
# open the datasets
with open('customer_profile.csv', 'rb') as f:
    customer_profile = f.read()
with open('transaction_data.csv', 'rb') as f:
    transaction_data = f.read()
```

```python
# decrypt the data
decrypt_customer_profile = fernet.decrypt(customer_profile)
decrypt_transaction_data = fernet.decrypt(transaction_data)
```

```python
# Overwrite the encrypted data with decryption
with open('customer_profile.csv', 'wb') as f:
    f.write(decrypt_customer_profile)
with open('transaction_data.csv', 'wb') as f:
    f.write(decrypt_transaction_data)
```
